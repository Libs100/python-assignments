## <font color = "pink" >  Part 1: Load and Inspect the Dataset



**Task:**

**Instructions:**

```python
import pandas as pd

# Load the dataset
# TODO: Read 'patient_data.csv' into a pandas DataFrame called 'data'
data = pd.read_csv('patient_data.csv')  

# Display the first 5 rows of the dataset
print(data.head(5))

# Count the number of patients
# TODO: Use a function to get the number of rows in 'data' and print it
print(f"Total number of patients: {len(data)}")
```

## <font color = "pink" > Task 2: Summarize the data

**Task**

Calculate the average (mean) WBC count.
Find the minimum and maximum WBC count.
Determine the number of unique conditions in the dataset.
Instructions:

Use data['WBC_Count'].mean() to compute the mean WBC count.
Use .min() and .max() to find the minimum and maximum WBC counts.
Use data['Condition'].nunique() to find the number of unique conditions.



```python
# Calculate basic statistics for WBC count
# Compute the mean WBC count and store it in 'mean_wbc'
mean_wbc = data['WBC_Count'].mean()
print(f"\nðŸ§¬ Average (Mean) WBC Count: {mean_wbc:.2f}")

# Compute the minimum WBC count and store it in 'min_wbc'
min_wbc = data['WBC_Count'].min()
print(f"ðŸ§ª Minimum WBC Count: {min_wbc}")

# Compute the maximum WBC count and store it in 'max_wbc'
max_wbc = data['WBC_Count'].max()
print(f"ðŸ”¬ Maximum WBC Count: {max_wbc}")

# Count unique conditions in the dataset
unique_conditions = data['Condition'].nunique()
print(f"\nðŸ“Š Number of Unique Conditions in the Dataset: {unique_conditions}")
```

## <font color = "pink" >Task 3: Filtering the data

**Task**

Extract patients with WBC count above 10,000.
Extract patients with the condition "Healthy".
Instructions:

Use boolean indexing to filter the DataFrame.
Store the filtered DataFrames in variables and display them.



```python
# Filter patients with WBC count > 10,000
high_wbc = data[data['WBC_Count'] > 10000]  # Filter rows where WBC_Count > 10,000
print("\nðŸ©¸ **Patients with WBC count above 10,000:**")
if not high_wbc.empty:
    print(high_wbc)
else:
    print("No patients found with WBC count above 10,000.")

# Filter patients with the condition 'Healthy'
healthy_patients = data[data['Condition'] == 'Healthy']  # Filter rows where Condition is 'Healthy'
print("\nðŸ©º **Patients with the condition 'Healthy':**")
if not healthy_patients.empty:
    print(healthy_patients)
else:
    print("No patients found with the condition 'Healthy'.")

```

## <font color = "pink" >Task 4: Counting the Data

```python
# Count patients by condition
condition_counts = data['Condition'].value_counts()  # Count the number of patients in each condition
print("\nðŸ“Š **Number of patients by condition:**")
print(condition_counts)

# Check if there are any conditions in the dataset
if not condition_counts.empty:
    print("\nðŸ¦  **Condition counts (detailed):**")
    for condition, count in condition_counts.items():
        print(f"ðŸ”¹ {condition}: {count} patients")
else:
    print("No conditions found in the dataset.")

```

## <font color ='pink'> # Task 5: Visualize Data

Task:

Create a bar chart showing the number of patients for each condition.
Create a scatter plot showing the relationship between age and WBC count.
Instructions:

Use matplotlib for plotting.
Customize the plots with titles and labels.






```python
import matplotlib.pyplot as plt

# Bar chart for number of patients by condition
plt.figure(figsize=(8, 6))  # Set the figure size
condition_counts.plot(kind='bar', color=['#FF6F61', '#6B5B95', '#88B04B', '#F7CAC9'], edgecolor='black')

plt.title('Number of Patients by Condition', fontsize=18, fontweight='bold', color='#333333')
plt.xlabel('Condition', fontsize=14, fontweight='bold', color='#555555')
plt.ylabel('Number of Patients', fontsize=14, fontweight='bold', color='#555555')
plt.xticks(rotation=0)  # Keep the x-axis labels straight
plt.grid(axis='y', linestyle='--', linewidth=0.7, color='gray', alpha=0.6)  # Add light horizontal gridlines
plt.show()


# Scatter plot of Age vs. WBC Count
plt.figure(figsize=(8, 6))  # Set the figure size
plt.scatter(x=data['Age'], y=data['WBC_Count'], c=data['WBC_Count'], cmap='viridis', s=100, alpha=0.8, edgecolors='black')

plt.title('Age vs. WBC Count', fontsize=18, fontweight='bold', color='#333333')
plt.xlabel('Age (in years)', fontsize=14, fontweight='bold', color='#555555')
plt.ylabel('WBC Count', fontsize=14, fontweight='bold', color='#555555')
plt.colorbar(label='WBC Count')  # Add colorbar to show WBC levels
plt.grid(linestyle='--', linewidth=0.7, color='gray', alpha=0.6)  # Add light gridlines
plt.show()
```
